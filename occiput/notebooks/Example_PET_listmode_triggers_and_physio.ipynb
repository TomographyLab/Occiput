# Physio and listmode triggers - Siemens Biograph mMR
This example shows how to load the Siemens physiologiacal data and listmode triggers. Such data includes respiratory and cardiac triggers, external signals, signals for the synchronization with between PET and MR, ..

- First, the example shows how to load the physiological data from the Dicom file generated by the Siemens scanners when physiological data is acquired (e.g., respiratory bellow, ECG, external signal). 

- Then the example shows how to load the triggers stored in the PET listmode data. These can include respiratory and cardiac signals, as well as signals for the synchronization with the MR scanner in the Biograph mMR.

## Load listmode triggers

When raw data is exported from the Siemens Biograph mMR through the console, the Siemens software exports 4 Dicom files. One of the files (whose name ends with 2.IMA) is just a wrapper of the raw listmode data stored in Interfile file format. The entire content of the listmode file is stored in a Dicom field. 

Occiput implements a utility function that extracts the listmode data from the Dicom file and stores it into an Interfile listmode file. Before loading the listmode data into Occiput, we have to convert the Dicom into listmode. 
You may wander why we don't read the binary data directly from the Dicom. The reason is that pydicom currently does not support streaming; we would have to laod the entire listmode file in RAM memory; but typically it's too large. 

Finally note that when the often the data is exported from the scanner directly in listmode format (i.e. without the Dicom wrapper). In fact there are two ways to export raw data from the scanner: through the console (producing the 4 Dicoms) and by copying the raw data from the hard drive of the reconstruction machine (raw listmode, no Dicom wrapper). 

Let's go.

```python
occiput.DataSources.FileSources.convert_listmode_dicom_to_interfile(basename+'2.IMA',basename+'listmode')
```

```python
scanner.listmode.load_gates(basename+'listmode.l',time_range_ms=[0,4000000])
gates = scanner.listmode.get_gates()
```

```python
time_gates_br_on  = gates['time'][gates['payload']==4096]
time_gates_br_off = gates['time'][gates['payload']==4097]
```

```python
figure(figsize=[18,4])
hold(0)
t1 = linspace(0,1000.0/50*len(physio['samples_breathing']),len(physio['samples_breathing']))
plot(t1,physio['samples_breathing'])
hold(1)
t2 = linspace(0,1000.0/50*len(physio['triggers_breathing']),len(physio['triggers_breathing']))
plot(t2,physio['triggers_breathing'],'r')
plot(time_gates_br_on,1000*ones(time_gates_br_off.shape),'xm')
plot(time_gates_br_off,1000*ones(time_gates_br_on.shape),'.g')
```

```python
figure(figsize=[18,4])
hold(0)

min_time = 0
max_time = 240000

sa = physio['samples_breathing']
tr = physio['triggers_breathing']
tsa = linspace(0,1000.0/50 *400/400 *len(sa),len(sa))
ttr = linspace(0,1000.0/50 *400/400 *len(tr),len(tr))
tgon  = time_gates_br_on
tgoff = time_gates_br_off
gon  = 1000*ones(tgon.shape)
goff = 1000*ones(tgoff.shape)

sa  = sa[(tsa>min_time)*(tsa<max_time)]
tsa = tsa[(tsa>min_time)*(tsa<max_time)]
tr  = tr[(ttr>min_time)*(ttr<max_time)]
ttr = ttr[(ttr>min_time)*(ttr<max_time)]
gon  = gon[(tgon>min_time)*(tgon<max_time)]
tgon = tgon[(tgon>min_time)*(tgon<max_time)]
goff  = goff[(tgoff>min_time)*(tgoff<max_time)]
tgoff = tgoff[(tgoff>min_time)*(tgoff<max_time)]

hold(1)
plot(tsa,sa)
plot(ttr,tr,'r')
plot(tgon,gon,'xm')
plot(tgoff,goff,'.g')
```

```python
time_gates_ecg_on  = gates['time'][gates['payload']==8192]
time_gates_ecg_off = gates['time'][gates['payload']==8193]
```
