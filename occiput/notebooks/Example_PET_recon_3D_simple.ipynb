```python
import occiput
```

```python
pet = occiput.Reconstruction.PET.PET_Static_Scan()
```

```python
pet.binning.display_geometry()
```

```python
activity = occiput.DataSources.Synthetic.Shapes.uniform_cylinder([128,128,128],axis=2)
```

```python
projection = pet.project_activity(activity)
```

```python
pet.set_prompts(projection)
```

```python
a = pet.osem_reconstruction(iterations=15, subset_size=64)
```

```python
subplot(1,3,1); imshow(a.data[:,64,:], cmap='gray')
subplot(1,3,2); imshow(a.data[64,:,:], cmap='gray')
subplot(1,3,3); imshow(a.data[:,:,64], cmap='gray')
```
