# Load motion information
Occiput can load motion information from MPRAGE MRI data acquired using volumetric navigators, from motion corrected EPI time series (including fMRI data) and from MRI endoscouts. 

```python
import occiput
```

### 1) Load vNav MPRAGE

```python
vnav = occiput.DataSources.FileSources.load_vnav_mprage('../../data/PET_data/scans/2014_04_16_keith/MR/EP_MOCO_NAV_SET_0034/')
```

```python
plt = vnav.plot_mean_displacement(figsize=[18,6],min_duration=0.0, extract_events_threshold=1.5)
```

### 3) Load motion information from MR magnetic field sensors (e.g., EndoScout)

```python
from occiput.DataSources.FileSources.MR_motion_sensors import Motion_Sensor
d = Motion_Sensor()
d.load_from_file('../../data/MR_data/MRI_reconstruction/res/multipos_150324192759.Res')
```

```python
d.plot_motion()
```

```python
events = d.extract_motion_events()
d.plot_mean_displacement()
```
